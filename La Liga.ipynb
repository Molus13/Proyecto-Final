```python
import matplotlib.pyplot as plt
```

```python
datos_jugadoras = [['Oshoala', [19, 2, 19, 0, 0]],
    ['Alba Redondo', [19, 5, 19, 5, 0]],
    ['Esther Gonzalez', [16, 4, 19, 2, 0]],
    ['Caroline Weir ', [13, 10, 19, 1, 0]],
    ['Racheal Kundananji ', [12, 3, 18, 5, 0]],
    
]

N = len(datos_jugadoras[0][1])
angles = np.linspace(0, 2*np.pi, N+1, endpoint=False)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(111, polar=True)

for jugadora in datos_jugadoras:
    jugadora[1] = np.concatenate((jugadora[1], [jugadora[1][0]]))

    ax.plot(angles, jugadora[1], 'o-', linewidth=2, label="Goles - Asistencias - " + jugadora[0])
    ax.fill(angles, jugadora[1], alpha=0.25)

plt.xticks(angles[:-1], ["Partidos Jugados", "Goles", "Asistencias", "Tarjetas Amarillas", "Tarjetas Rojas"])
plt.legend(loc='upper right', bbox_to_anchor=(0.0, 0.0))

plt.show()
   
```

```python
datos_jugador = ['Oshoala', [19, 2, 19, 0, 0]]

N = len(datos_jugador[1]) - 2 

angles = np.linspace(0, 2*np.pi, N+1, endpoint=False)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(111, polar=True)

datos_jugador[1] = np.concatenate((datos_jugador[1][:N], [datos_jugador[1][0]])) 

ax.plot(angles, datos_jugador[1], 'o-', linewidth=2, label="Goles - Asistencias - Partidos Jugados")
ax.fill(angles, datos_jugador[1], alpha=0.25)

plt.xticks(angles[:-1], ["Goles", "Asistencias", "Partidos Jugados"])
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()


```
