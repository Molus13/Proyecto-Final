```python
datos_jugadoras = [['Oshoala', [19, 2, 19]],
    ['Alba Redondo', [19, 5, 19]],
    ['Esther Gonzalez', [16, 4, 20]],
    ['Caroline Weir ', [13, 10, 15]],
    ['Racheal Kundananji ', [12, 3, 20]],
    
]

N = len(datos_jugadoras[0][1])
angles = np.linspace(0, 2*np.pi, N+1, endpoint=False)

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111, polar=True)

for jugadora in datos_jugadoras:
    jugadora[1] = np.concatenate((jugadora[1], [jugadora[1][0]]))

    ax.plot(angles, jugadora[1], 'o-', linewidth=2, label="Goles - Asistencias - " + jugadora[0])
    ax.fill(angles, jugadora[1], alpha=0.25)

plt.xticks(angles[:-1], ["Partidos Jugados", "Goles", "Asistencias"])
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
   
```
